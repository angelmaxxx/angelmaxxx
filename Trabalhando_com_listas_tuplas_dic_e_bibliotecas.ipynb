```python
# Tuplas (criando tuplas, processo semelhante ao de criação de listas, mas
# ao invés de colchetes se usam parenteses)
'''
A diferença entre lista e tuplas é a mutabilidade, ou seja, as listas são
mutáveis, o que significa que você pode modificar, adicionar ou remover
elementos depois de criá-las. Por outro lado, as tuplas são imutáveis, o que
significa que, uma vez criadas, não é possível alterar seus elementos. Se você
precisar modificar uma tupla, terá que criar uma nova com os elementos
desejados. By Bianka.... apud Chat GPT
'''
nomest = ('Douglas','Daniela','Pedro','Maria','Eduardo','Ester')
idadest = tuple(idades)
alturast = tuple(alturas)
pesost = tuple(pesos)
```

```python
# Transformando uma tupla em uma lista
lista = list(pesost)
type(lista)
```

```python
# Criando dicionários
'''
Um dicionário é uma estrutura compsta por um ou mais pares de dados no sistema
de chave e valor, exemplo
dicionario = {chave1:valor1, chave2:valor2, .... chaven:valorn}
'''
id1 = {'nome':'Douglas', 'idade':45, 'altura':1.85, 'peso':70}
id2 = {'nome':'Daniela','idade':7,'altura':1.23,'peso':22}
id3 = {'nome':'Pedro','idade':65,'altura':1.75,'peso':87}
id4 = {'nome':'Maria','idade':64,'altura':1.67,'peso':64}
id5 = {'nome':'Eduardo','idade':37,'altura':1.82,'peso':96}
id6 = {'nome':'ester','idade':37,'altura':1.73,'peso':68}
```

SCIPY

```python
#Trabalhando com a parte estatística do SciPy (estatística descritiva)

# média aritmética no SciPy
print(f' A idade média é {sp.mean(idades)} anos')

# mediana no SciPy
print(f' A idade mediana é {sp.median(idades)} anos')

# desvio padrão no SciPy (populacional)
print(f' O desvio padrão populacional é {sp.std(idades):.2f} anos')

# desvio padrão no SciPy (amostral)
print(f' O desvio padrão amostral é {sp.std(idades,ddof=1):.2f} anos')

# Variância no SciPy (populacional)
print(f' A Variância populacional é {sp.var(idades):.2f} anos')

# Variância no SciPy (amostral)
print(f' A Variância amostral é {sp.var(idades,ddof=1):.2f} anos')

# Q1 (primeiro quartil) no SciPy
print(f' O valor do Q1 é {sp.quantile(idades, 0.25):.2f} anos')

# Q2 (segundo quartil) no SciPy
print(f' O Valor do Q2 é {sp.quantile(idades, 0.5):.2f} anos')

# Q3 (terceiro quartil) no SciPy
print(f' O Valor do Q3 é {sp.quantile(idades, 0.75):.2f} anos')

# Valor máximo no SciPy
print(f' O Valor máximo é {sp.amax(idades):.2f} anos') # ou usando quantile
print(f' O Valor máximo é {sp.quantile(idades,1):.2f} anos') # ou usando quantile

# Valor mínimo no SciPy
print(f' O Valor mínimo é {sp.amin(idades):.2f} anos') # ou usando quantile
print(f' O Valor mínimo é {sp.quantile(idades,0):.2f} anos') # ou usando quantile

# Amplitude total
print(f' A amplitude total é {sp.amax(idades)-sp.amin(idades):.2f}')

# Amplitude interquartílica AIQ
print(f' A AIQ é {sp.quantile(idades, 0.75)-sp.quantile(idades, 0.25):.2f}')

# Limite inferior do boxplot
print(f' O limite inferior do box plot é {(sp.quantile(idades, 0.25))-((sp.quantile(idades, 0.75)-sp.quantile(idades, 0.25))*1.5):.2f}')

# Limite superior do boxplot
print(f' O limite superior do box plot é {(sp.quantile(idades, 0.75))+((sp.quantile(idades, 0.75)-sp.quantile(idades, 0.25))*1.5):.2f}')
```

```python
def eda(medidas):
  sd = list(medidas.iloc[2])
  sd.pop(0)
  q1 = list(medidas.iloc[4])
  q1.pop(0)
  q3 = list(medidas.iloc[6])
  q3.pop(0)
  min = list(medidas.iloc[3])
  min.pop(0)
  max = list(medidas.iloc[7])
  max.pop(0)
  var = []
  at = []
  aiq = []
  ls = []
  li = []
  for i in range(len(sd)):
    var.append(sd[i]**2)
    at.append(max[i]-min[i])
    aiq.append(q3[i]-q1[i])
    ls.append(q3[i]+(1.5*aiq[i]))
    li.append(q1[i]-(1.5*aiq[i]))
  var.insert(0, 'variância')
  at.insert(0, 'AT')
  aiq.insert(0, 'AIQ')
  ls.insert(0, 'LS')
  li.insert(0, 'LI')
  medidas.loc[9] = var
  medidas.loc[10] = at
  medidas.loc[11] = aiq
  medidas.loc[12] = ls
  medidas.loc[13] = li
  display(medidas)
```
